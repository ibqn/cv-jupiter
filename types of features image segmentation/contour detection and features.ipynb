# Finding Contours

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv

%matplotlib inline
```

```python
img_hand = cv.imread('images/raised_hand2.jpg')
img_hand = cv.cvtColor(img_hand, cv.COLOR_BGR2RGB)

plt.imshow(img_hand)
plt.show()
```

## Produce a binary image for finding contours


```python
img_gray = cv.cvtColor(img_hand, cv.COLOR_RGB2GRAY)

# create a binary thresholded image
# ignoring retval output parameter
_, img_binary = cv.threshold(img_gray, thresh=230, maxval=255, type=cv.THRESH_BINARY_INV)

plt.imshow(img_binary, cmap='gray')
plt.show()
```
