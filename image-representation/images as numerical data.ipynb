# Image as Grid of Pixels

## Import resources

```python
import numpy as np
from matplotlib import(
    image as mpim,
    pyplot as plt,
)
import cv2 as cv

%matplotlib inline
```

## Read and display images

```python
img = mpim.imread('images/waymo_car.jpg')
print(f'image dimensions: {img.shape}')
plt.imshow(img)
plt.show()
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
tiny_image = np.array([
    [0, 20, 30, 150, 120],
    [200, 200, 250, 70, 3],
    [50, 180, 85, 40, 90],
    [240, 100, 50, 255, 10],
    [30, 0, 75, 190, 220]
])

plt.matshow(tiny_image, cmap='gray')
plt.show()
```
