# Blue Screen

```python
from matplotlib import(
    pyplot as plt,
    image as mpim,
)
import cv2 as cv
import numpy as np
```

```python
img = mpim.imread('images/pizza_bluescreen.jpg')
print(f'image shape {img.shape}')
h, w = img.shape[:2]
plt.imshow(img)
plt.show()
```

## Create a mask of an image

```python
lower_blue = np.array([0, 0, 100])
upper_blue = np.array([255, 65, 255])

mask = cv.inRange(img, lower_blue, upper_blue)

plt.imshow(mask, cmap='gray')
plt.show()
```

## Apply mask to an image

```python
img_masked = np.copy(img)
img_masked[mask != 0] = [0, 0, 0]
plt.imshow(img_masked)
plt.show()
```

```python
img_bg = mpim.imread('images/space_background.jpg')
print(f'shape of the background image: {img_bg.shape}')

img_cropped_bg = np.copy(img_bg[:h,:w])
img_cropped_bg[mask == 0] = [0, 0, 0]

plt.imshow(img_cropped_bg)
plt.show()
```

## Overlay both images

```python
img_overlay = img_cropped_bg + img_masked

plt.imshow(img_overlay)
plt.show()
```
