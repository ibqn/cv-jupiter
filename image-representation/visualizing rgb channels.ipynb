# RGB Color space

```python
import numpy as np
from matplotlib import(
    image as mpim,
    pyplot as plt,
)
import cv2 as cv

%matplotlib inline
```

## Reduce the size of an image

The `cv.imread` function  from opencv opens an image in BGR color space, while `mpim.imread` opens it in RGB color space.

```python
img = cv.imread('images/wa_state_highway.jpg', cv.IMREAD_COLOR)

h, w = img.shape[0:2]
hh, ww = h//4, w//4

print(f'{(h, w)} -> {(hh, ww)}')

img2 = cv.resize(img, dsize=(ww, hh))

# remove huge image, as we do not need it any longer
del img 

# convert image from BGR to RGB so we can show it
plt.imshow(img2[:,:,::-1])
plt.show()

cv.imwrite('images/wa_state_highway-small.jpg', img2)
```

```python
img = mpim.imread('images/wa_state_highway-small.jpg')
plt.imshow(img)
plt.show()
```


## RGB channels

Visualize the levels of each color channel. Pay close attention to the traffic signs!


```python
# isolate RGB channels
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]

img_united = np.hstack((r, g, b))

# visualize the individual color channels
plt.figure(figsize=(18, 4))
plt.title('Red channel | Green channel | Blue channel')
plt.imshow(img_united, cmap='gray')
plt.show()
```
