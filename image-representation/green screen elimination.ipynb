```python
from matplotlib import(
    pyplot as plt,
    image as mpim,
)
import cv2 as cv
import numpy as np
```

```python
img = cv.imread('images/car_green_screen2.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

## Background elimination in RGB color space

Visualize the threshold of the green color range defined in the previous, consistent green color case.

```python
# define our color selection boundaries in RGB values
lower_green = np.array([0, 45, 0]) 
upper_green = np.array([100, 255, 100])

# define the masked area
mask = cv.inRange(img, lower_green, upper_green)
mask = cv.bitwise_not(mask)

# mask the image to let the car show through
img_masked = np.copy(img)
img_masked[mask == 0] = [110, 110, 110]
#img_masked = cv.bitwise_and(img, img, mask=mask)

plt.imshow(img_masked)
plt.show()
```

## Background elimination in HSV color space

Visualize the green color range for background removal.

```python
img_hsv = cv.cvtColor(img, cv.COLOR_RGB2HSV)

lower_hue = np.array([0, 45, 5]) 
upper_hue = np.array([100, 255, 255])

# define the masked area
mask = cv.inRange(img, lower_hue, upper_hue)

# mask the image to let the car show through
img_mask = np.copy(img_hsv)

img_mask = cv.cvtColor(img_mask, cv.COLOR_HSV2RGB)
img_mask[mask != 0] = [110, 110, 110]

plt.imshow(img_mask)
plt.show()
```
