```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv

%matplotlib inline
```

```python
img_brain = cv.imread('images/brain_MR.jpg', cv.IMREAD_GRAYSCALE)

plt.imshow(img_brain, cmap='gray')
plt.show()
```

# Canny edge detection

```python
wide = cv.Canny(img_brain, threshold1=30, threshold2=100)
tight = cv.Canny(img_brain, threshold1=200, threshold2=240)

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.imshow(wide, cmap='gray')
ax2.imshow(tight, cmap='gray')

plt.show()
```

```python
img_flower = cv.imread('images/sunflower.jpg', cv.IMREAD_GRAYSCALE)

lower_thresh = 130
upper_thresh = 160

edges_flower = cv.Canny(img_flower, lower_thresh, upper_thresh)

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

ax1.imshow(img_flower, cmap='gray')
ax2.imshow(edges_flower, cmap='gray')

plt.show()
```
